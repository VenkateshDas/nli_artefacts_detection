```python
def annotate_bars(ax):
    """Annotate bars with their value."""
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2.,
                height + 0.01,  # Adding a small offset to position the text above the bar
                '{:1.3f}'.format(height),
                ha="center", rotation=90)

def plot_model_performance_facet_grid(full_context, hypothesis_only, majority_baseline, model_name):

  df_full_context = pd.DataFrame(full_context)
  df_full_context["Model"] = f"{model_name} Full-Context"
  df_hypothesis_only = pd.DataFrame(hypothesis_only)
  df_hypothesis_only["Model"] = f"{model_name} Hyp-Only"
  df_majority_baseline = pd.DataFrame(majority_baseline)
  df_majority_baseline["Model"] = "Majority Baseline"

  df_combined = pd.concat([df_full_context, df_hypothesis_only, df_majority_baseline])
  df_combined_melt = df_combined.melt(id_vars=["Features", "Model"], var_name="Year", value_name="Performance")

  # Setting up a larger Facet Grid with bar charts
  g = sns.FacetGrid(df_combined_melt, col="Features", col_wrap=3, height=7, sharey=True)

  # Adding bars for each model
  g.map_dataframe(sns.barplot, x="Year", y="Performance", hue="Model", palette="tab20c", lw=2.5)

  for ax in g.axes.flat:
      annotate_bars(ax)

  # Enhancing the visualization
  g.fig.suptitle(f'{model_name} COLIEE Models Performance', fontsize=16, y=1.03)
  g.set_titles("Feature = {col_name}")
  g.set_xlabels("")
  g.set_ylabels("Accuracy")
  g.add_legend(loc="upper right", bbox_to_anchor=(1.05, 1.05))
  plt.tight_layout()
  plt.show()
```

```python
def annotate_bars(ax):
    for p in ax.patches:
        ax.annotate(f"{p.get_height():.2f}",
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center',
                    xytext=(0, 5),
                    textcoords='offset points',
                    fontsize=9, rotation=0)

def plot_individual_yearly_comparison(dataframe, context_type="full-context"):
    # Exclude majority_baseline
    dataframe = dataframe[dataframe["Model"] != "Majority Baseline"]

    # Filter dataframe based on context_type
    if context_type == "full-context":
        dataframe = dataframe[dataframe["Model"].str.contains("Full-Context")]
    elif context_type == "hyp-only":
        dataframe = dataframe[dataframe["Model"].str.contains("Hyp-Only")]

    # Loop through each year and create a separate plot for each
    for year in ["2018", "2019", "2020", "2021", "2022"]:
        # Define the plot size
        plt.figure(figsize=(15, 8))

        # Filter dataframe for the specific year
        yearly_data = dataframe[dataframe["Year"] == year]

        # Create the barplot
        ax = sns.barplot(data=yearly_data, x="Model", y="Performance", hue="Features", palette="tab20c")

        # Annotate bars with their respective values
        annotate_bars(ax)

        # Set the title for the plot
        plt.title(f"Model Performance Comparison for Year {year}")

        # Rotate x-labels for better visibility
        plt.xticks(rotation=45)

        # Display the plot
        plt.tight_layout()
        plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1))
        plt.show()

```

```python
# Call the function to visualize the individual yearly comparison for full-context models
plot_individual_yearly_comparison(combined_df, context_type="full-context")

# Call the function to visualize the individual yearly comparison for hyp-only models
plot_individual_yearly_comparison(combined_df, context_type="hyp-only")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_boxplots(combined_df, context_type="Full-Context"):
    """
    Plots box plots of model performances for each year.
    :param combined_df: The combined dataframe of performances.
    :param context_type: Either "Full-Context" or "Hyp-Only".
    """

    # Filtering out the majority_baseline and selecting the desired context
    filtered_df = combined_df[~combined_df["Model"].isin(["Majority Baseline"]) & combined_df["Model"].str.contains(context_type)]

    plt.figure(figsize=(15, 8))

    # Creating box plots
    sns.boxplot(data=filtered_df, x="Year", y="Performance", hue="Model", palette="PuBuGn")

    plt.title(f"Box Plot of Model Performances ({context_type}) Across Years")
    plt.ylabel("Accuracy")
    plt.xlabel("Year")
    plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1))
    plt.tight_layout()
    plt.show()


```

```python
# To plot the box plots for Full-Context models:
plot_boxplots(combined_df, context_type="Full-Context")
```

```python
# To plot the box plots for Hyp-Only models:
plot_boxplots(combined_df, context_type="Hyp-Only")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_violin(combined_df, context_type="Full-Context"):
    """
    Plots violin plots combined with swarm plots of model performances for each year.
    """
    # Filtering out the majority_baseline and selecting the desired context
    filtered_df = combined_df[~combined_df["Model"].isin(["Majority Baseline"]) & combined_df["Model"].str.contains(context_type)]

    plt.figure(figsize=(15, 10))

    # Creating violin plots
    sns.violinplot(data=filtered_df, x="Year", y="Performance", hue="Model", palette="PuBuGn", inner=None, alpha=0.7)

    # Adding swarm plots
    # sns.swarmplot(data=filtered_df, x="Year", y="Performance", hue="Model", dodge=True, palette="dark", size=6)

    plt.title(f"Violin and Swarm Plot of Model Performances ({context_type}) Across Years")
    plt.ylabel("Accuracy")
    plt.xlabel("Year")
    plt.legend(loc="upper right", bbox_to_anchor=(1.21, 1))
    plt.tight_layout()
    plt.show()
```

```python
# To plot the violin and swarm plots for Full-Context models:
plot_violin(combined_df, context_type="Full-Context")

# To plot the violin and swarm plots for Hyp-Only models:
plot_violin(combined_df, context_type="Hyp-Only")
```

```python
def calculate_difference(full_context_data, hyp_only_data):
    """
    Calculate the performance difference between full context and hypothesis only models.
    """
    df_full = pd.DataFrame(full_context_data).set_index("Features")
    df_hyp = pd.DataFrame(hyp_only_data).set_index("Features")

    # Subtracting the hypothesis only performance from full context performance
    difference = df_full.subtract(df_hyp).reset_index()

    return difference
def annotate_bars(ax):
    """
    Annotate the bars with the bar height (rounded to 4 decimal places).
    """
    for p in ax.patches:
        # Get the height of the bar (which represents the difference)
        height = p.get_height()

        # Position the text in the middle of the bar, slightly offset from the top
        ax.annotate(f'{height :.2f}',
                    (p.get_x() + p.get_width() / 2., p.get_y() + height),
                    ha='center', va='center',
                    xytext=(0, 9),
                    textcoords='offset points', rotation=0)

def plot_difference_bar_graph(difference_df, model_name):
    """
    Plot a bar graph showing the performance difference for each year.
    """
    # Melt the dataframe to long format
    df_melted = difference_df.melt(id_vars=["Features"], var_name="Year", value_name="Difference")

    plt.figure(figsize=(15, 7))

    # Plotting the bar graph
    ax = sns.barplot(data=df_melted, x="Year", y="Difference", hue="Features", palette="tab20c")

    # Annotate bars with the difference values
    annotate_bars(ax)

    # Setting title and labels
    plt.title(f"Performance Difference Between Full Context and Hyp-Only for {model_name}")
    plt.axhline(0, color='black',linewidth=1)
    plt.ylabel("Performance Difference")
    plt.xlabel("Year")
    plt.legend(loc="upper right", bbox_to_anchor=(1.22, 1))
    plt.tight_layout()
    plt.show()

# Calculate the difference and plot the bar graph for BERT model as an example
bert_difference = calculate_difference(bert_base_full_context, bert_base_hypothesis_only)
plot_difference_bar_graph(bert_difference, "BERT Base")

# Similarly, the same can be done for other models
```
